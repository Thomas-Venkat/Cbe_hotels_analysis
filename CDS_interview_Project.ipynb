```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.spatial import ConvexHull
import sqlite3
```

```python
conn = sqlite3.connect("/content/db.sqlite3")
cur = conn.cursor()
data = cur.execute("SELECT * FROM restAPI_hoteldetail")

# f = open('hoteldetails.csv', 'w')
# for row in data:
#   print(row)
# f.close()  
import csv

colnames = ['id', 'name', 'address', 'location', 'city', 'state', 'country']
allrows = [row for row in data]

filename = "hoteldetails.csv"

with open(filename, 'w') as csvfile:
  csvwriter = csv.writer(csvfile)
  csvwriter.writerow(colnames)
  csvwriter.writerows(allrows)
```

```python
df = pd.read_csv("/content/hoteldetails.csv")
```

```python
df.shape
```

```python
df.dropna(inplace=True)
new = df['location'].str.split(",", n = 1, expand=True)
df['lat'] = pd.to_numeric(new[0])
df['long'] = pd.to_numeric(new[1])
df.drop(columns = ['location'], inplace=True)
```

```python
df['restaurants'] = df['lat'].map(df['lat'].value_counts())
```

```python
new_df = df.drop_duplicates(subset="lat")
new_df
```

```python
new_df = new_df.drop(['id', 'name', 'address', 'city', 'state', 'country'], axis=1)
```

```python
new_df
```

```python
K_clusters = range(1, 7)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
Y_axis = new_df[['lat']]
X_axis = new_df[['long']]
score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]
# Visualize
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init ='k-means++')
y_kmeans = kmeans.fit_predict(X_axis)
new_df['clusters'] = y_kmeans +1
```

```python
import folium
from folium.plugins import MarkerCluster
m = folium.Map(location=[new_df['lat'].mean(), new_df['long'].mean()],
               tiles = 'CartoDB positron', zoom_start=12)
```

```python
marker_cluster = MarkerCluster().add_to(m)


folium.Marker(
    location=[11.019508,	76.967260],
    popup="Gandhipuram",
    icon=folium.Icon(color="green", icon="ok-sign"),
).add_to(marker_cluster)

folium.Marker(
    location=[11.036491,	76.954729],
    popup="Sanganoor",
    icon=folium.Icon(color="red", icon="remove-sign"),
).add_to(marker_cluster)

folium.Marker(
    location=[11.012462,	76.949312],
    popup="RS Puram",
    icon=folium.Icon(color="blue", icon="ok-sign"),
).add_to(marker_cluster)

m
```
